Print the password length first then create function validating if the password has more than eight characters. If the password fulfills the above condition and has more then eight character print 'valid password' otherwise print invalid password and ask the user to enter the password again.

````python
user_password = input('Your Password:')
print(f'Your password length is: {len(user_password)}')

def pass_validity(user_password) -> bool:
    return len(user_password) >= 8

is_valid = False  # Boolean flag to track validity. false so loops runs at least once.

while not is_valid:
    user_password = input('Your Password: ')
    print(f'Your password length is: {len(user_password)}')
    
    is_valid = pass_validity(user_password) # calling function
    
    if is_valid == True:
        print("Your Password is valid.")
    else:
        print("Invalid Password. Must be at least 8 characters.")
          
````

2. Ask user for two inputs of a name and surname Create a function generating a special secret word compris the first three letters of the username and last two characters of the users surname The function returns the special word

````python
def generate_secret_word(name, surname):
    part1 = name[:3]
    part2 = surname[-2:]
    return part1 + part2

def main():
    print("Secret Word Creator")
    name = input("Enter your first name: ").strip()
    surname = input("Enter your surname: ").strip()

    if len(name) < 3 or len(surname) < 2:
        print("Name must be at least 3 letters and surname at least 2 letters.")
    else:
        secret_word = generate_secret_word(name, surname)
        print(f"Your special secret word is: {secret_word}")

if __name__ == "__main__":
    main()
````

3. Create a function taking one parameters for a password. The function would convert each character of the password into corresponding ASCII numbers and return one string composed of the ASCII numbers instead

````python
def password_to_ascii(password):
    ascii_string = ''
    for char in password:
        ascii_string += str(ord(char))
    return ascii_string

def main():
    password = input("Enter your password: ")
    converted = password_to_ascii(password)
    print(f"Your ASCII-encoded password is: {converted}")

if __name__ == "__main__":
    main()
````

4. Create a function (or set of functions ) taking one parameter for a password.

The function (s) will print True if

  - the password contains at least one upper case letter,   
  - the password contains at least one lower case letter,
  - the password contains at least one digit and one special character like for example ‚Äò@‚Äô.
- Extra (check if the password contains space. If so, print a suitable error message)

````python
def check_password(password):
    if ' ' in password:
        print("Error: Password should not contain spaces.")
        return

    has_upper = any(char.isupper() for char in password)
    has_lower = any(char.islower() for char in password)
    has_digit = any(char.isdigit() for char in password)
    has_special = any(char in '@#$%^&*!?' for char in password)

    print(f"Contains uppercase letter: {has_upper}")
    print(f"Contains lowercase letter: {has_lower}")
    print(f"Contains digit: {has_digit}")
    print(f"Contains special character (@#$%^&*!?): {has_special}")

def main():
    password = input("Enter your password: ")
    check_password(password)

if __name__ == "__main__":
    main()
````

5. Create a function validating a student number. The function takes one parameter for a student number and returns True if the student number starts with ‚ÄúM00‚Äù, and the student number has 9 characters

````python
def is_valid_student_number(student_number):
    return student_number.startswith("M00") and len(student_number) == 9

def main():
    student_number = input("Enter your student number: ").strip()
    if is_valid_student_number(student_number):
        print("Valid student number.")
    else:
        print("Invalid student number. It must start with 'M00' and be exactly 9 characters long.")

if __name__ == "__main__":
    main()
````

7. Create a function generating an email for an organization: Ask the user for his or her initials. Make sure that there is no whitespace added to the initials Before adding the initials make sure that there are lower case letters Concatenate the initial with a randomly generated three-digit number and the ‚Äò@live.mdx.ac.uk‚Äô. Return the new email.

````python
import random

def generate_email(initials):
    # Remove whitespace and convert to lowercase
    clean_initials = initials.replace(" ", "").lower()

    # Generate a random 3-digit number
    random_number = random.randint(100, 999)

    # Concatenate to form the email
    email = f"{clean_initials}{random_number}@live.mdx.ac.uk"
    return email

def main():
    initials = input("Enter your initials: ").strip()
    if not initials:
        print("Error: Initials cannot be empty.")
        return

    email = generate_email(initials)
    print(f"Your generated email is: {email}")

if __name__ == "__main__":
    main()
````


8. Write a program that asks the user to enter two numbers and divides them. Use try-except to catch any ZeroDivisionError and ValueError that might occur (e.g., division by zero or invalid input).

````python
def divide_numbers():
    try:
        num1 = float(input("Enter the first number: "))
        num2 = float(input("Enter the second number: "))
        result = num1 / num2
        print(f"Result: {num1} √∑ {num2} = {result}")
    except ZeroDivisionError:
        print("Error: You can't divide by zero.")
    except ValueError:
        print("Error: Please enter valid numbers.")

def main():
    divide_numbers()

if __name__ == "__main__":
    main()
````
