### Threshold 2 Multiplication Quiz

```python
#importing the random package
import random
number1 = random.randint(0,9)
number2 = random.randint(0,9)

print(f"What is {number1} * by {number2}")

#asking student for input with input validation
try:
    std_answer = int(input("Enter your answer here:"))
except ValueError:
    print("Enter a number like 3 or 5 (digits only)")

correct_ans = number1*number2

# using ternary operator and nested f string
print(f"{'Correct Answer' if std_answer == correct_ans else f'Incorrect. The correct answer is: {correct_ans}'}")
```

### Threshold 4 BMI

```python
user_weight_lbs = int(input("Enter weight in pounds:"))
user_height_inches = int(input("Enter user height in inches:"))


kg = user_weight_lbs * 0.45359237
meters = user_height_inches * 0.0254

bmi = kg/(meters*meters)
print(f"Your BMI is: {bmi:.2f}")

if bmi < 18.5:
    print("Underweight")
elif bmi >= 18.5 and bmi <= 24.9:
    print("Normal")
elif bmi >= 25 and bmi <= 29.9:
    print("Overweight")
elif bmi >=30:
     print("Obese")
```

### Threshold 5 Student Mark Analysis

```python
# @audit ask maam about this
marks = []

for i in range(5):  # marks for 5 subjects for 1 student
    while True:
        try:
            mark = int(input(f"Enter mark for student {i + 1}: "))
            marks.append(mark)
            break
        except ValueError:
            print("Please enter a valid number (digits only).")

print("Max marks are:", max(marks))
print("Least marks are:", min(marks))

total = 0
for mark in marks:
    total = total + mark

average = total / len(marks)
print(f"Average mark: {average:.2f}")

if average >= 50:
    print("passed")
else:
    print("failed")
```

### Threshold 11 Math Package using the sq root module

### Typical 15 Highest student score

```python
names = []
scores = []

try: 
    std_number = int(input("User enter the number of students:"))
except ValueError:
    print("enter a valid whole number")
for t in range(std_number):
    name = input(f'Enter student name {t + 1}')
    score = float(input(f'Enter student score for: {name}'))
    
    names.append(name)
    scores.append(score)

# Works cuz a parallel list is used as score values match position of std names
highest_mark = max(scores)  # Find the highest score
index = scores.index(highest_mark)  # Find the index of that score
print(f"The highest mark is {highest_mark}, scored by {names[index]}")
```
