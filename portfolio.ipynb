### Week 2 Typical 15 --- Find the highest score amongst the number of students entered:

```python
# MISIS: M01099037
# NAME: Muhammad Asad Ahmad.

names = []
scores = []

try:
    std_number = int(input("User enter the number of students: "))
except ValueError:
    print("enter a valid whole number")

for student_index in range(std_number):
    name = input(f'Enter student name {student_index + 1}: ')
    score = float(input(f'Enter student score for: {name}: '))

    # Store student name and score in respective lists
    names.append(name)
    scores.append(score)

# Find the highest score and the corresponding student name
highest_mark = max(scores)
index = scores.index(highest_mark)

print(f"Total number of students is: {std_number}")
print(f"The highest mark is {highest_mark}, scored by {names[index]}")
```

### Week 4 Excellent 17 Decimal to Binary.

```python
# MISIS: M01099037
# NAME: Muhammad Asad Ahmad.

def decimalToBinary(value):
    """
    Summary: Converts Decimal integer to Binary.

    Args: (value) integer value taken from user.

    Returns:
        str: binary, which is the converted decimal integer.
    """
    
    if value == 0:
        return "0"
    
    else:
        binary = "" # Empty String
        while value > 0:
            remainder = value % 2 # Modulo Division (gives remainder in division)
            binary = str(remainder) + binary # Prepending remainder to front of binary.
            value = value // 2 # Floor or Integer division (rounds down result to the nearest whole number)
        return binary

try:
    user_input = int(input("Enter Decimal Number here: "))
    print("You Entered Decimal Number:", user_input)
    print("Binary representation:", decimalToBinary(user_input))
except ValueError:
    print("Enter a valid whole number.")
```

### Week 5 Excellent Lesbrown text file word counter

```python
# MISIS: M01099037
# NAME: Muhammad Asad Ahmad.

from collections import Counter # Word Counter
import string # Special Characters !"#$%&'()*+,-./:;<=>?@[\]^_`{|}~

try: 

    with open("Les_Brown.txt", "r", encoding="utf-8") as file:
        text = file.read()

    # Clean and split the text into words.
    text = text.lower()
    clean_text = "" # empty string  

    for char in text:
        if char not in string.punctuation:
            clean_text += char  

    words = clean_text.split() # gives list of individual words.    

    # Word count frequencies stored as a dictionary.
    word_counts = Counter(words)    

    # Write the word counts to a new file word_count.txt
    with open("word_count.txt", "w", encoding="utf-8") as output_file:
        for word, count in word_counts.items(): # Dictionary method used.
            output_file.write(f"{word}: {count}\n")

except FileNotFoundError:
    print("The specific file in quesyion wasn't found. Oopsies.")

except Exception as other_errors:
    print(other_errors)
```
